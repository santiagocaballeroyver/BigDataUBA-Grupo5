```python
# Reemplazamos los datos correspondientes a NS/NC a NaN
def reemplazar_nsnc_por_nan(df):
    df_clean = df.copy()

    # Variables categóricas donde 9 es NS/NC
    vars_cat = ['CAT_OCUP', 'EMPLEO', 'SECTOR', 'CH07', 'CH08']
    for col in vars_cat:
        df_clean[col] = df_clean[col].replace(9, np.nan)

    # Códigos específicos de NS/NC del INDEC
    df_clean['PP3E_TOT'] = df_clean['PP3E_TOT'].replace(999, np.nan)     # Horas
    df_clean['PP04C'] = df_clean['PP04C'].replace(99, np.nan)            # Tamaño empresa
    df_clean['PP04D_COD'] = df_clean['PP04D_COD'].replace(99999, np.nan) # Ocupación
    df_clean['ESTADO'] = df_clean['ESTADO'].replace(0, np.nan)           # Entrevista no realizada

    # Valores "0" que significan "No corresponde"
    df_clean.loc[df_clean['CAT_OCUP'] == 0, 'CAT_OCUP'] = np.nan
    df_clean.loc[df_clean['PP07H'] == 0, 'PP07H'] = np.nan

    # Ingresos: Cualquier valor <= 0 (incluyendo el -9 de NS/NC) pasa a NaN
    df_clean.loc[df_clean['P21'] <= 0, 'P21'] = np.nan
    df_clean.loc[df_clean['P47T'] <= 0, 'P47T'] = np.nan

    # ---------------------------------------------------------
    # Limpieza outliers:
    # ---------------------------------------------------------

    # Nos quedamos estrictamente con la Población en Edad de Trabajar (PET)
    df_clean = df_clean[df_clean['CH06'] >= 15]

    # Límite físico humano de horas trabajadas (tope conservador)
    df_clean.loc[df_clean['PP3E_TOT'] > 120, 'PP3E_TOT'] = np.nan

    # Trimming (recorte) del 1% superior de la distribución de ingresos
    if df_clean['P21'].notna().sum() > 0:
        q99_p21 = df_clean['P21'].quantile(0.99)
        df_clean.loc[df_clean['P21'] > q99_p21, 'P21'] = np.nan

    if df_clean['P47T'].notna().sum() > 0:
        q99_p47t = df_clean['P47T'].quantile(0.99)
        df_clean.loc[df_clean['P47T'] > q99_p47t, 'P47T'] = np.nan

    return df_clean

# Aplicamos la limpieza a las bases
T424_v = reemplazar_nsnc_por_nan(T424_v)
T425_v = reemplazar_nsnc_por_nan(T425_v)
```

Con el objetivo de garantizar la robustez del análisis estadístico, se procedió a la detección y depuración de outliers e inconsistencias en las variables de interés. En primer lugar, se aplicó un filtro de edad para trabajar exclusivamente con la Población en Edad de Trabajar (≥15 años). Posteriormente, para las variables de ingresos (P21 y P47T), se eliminaron valores iguales o menores a cero y se aplicó un recorte en el percentil 99 de la distribución para mitigar el sesgo introducido por valores extremos que no reflejan la realidad promedio de los trabajadores. Finalmente, respecto a la variable de horas trabajadas (PP3E_TOT), se descartaron registros superiores a las 120 horas semanales por resultar físicamente inviables, asegurando así la integridad de los estimadores salariales y la comparabilidad de los resultados entre los períodos analizados.

## Parte III: Conociendo la Informalidad Laboral

Siguiendo la definición de Maurizio & Monsalvo (2021), la categoría **Informal self-employed** corresponde a trabajadores por cuenta propia no profesionales y trabajadores familiares sin remuneración que se desempeñan en el sector informal.

Para construir el indicador se utilizaron las siguientes variables del diccionario de la EPH:

- **CAT_OCUP** (Categoría ocupacional): se seleccionaron los valores 2 (cuenta propia) y 4 (trabajador familiar sin remuneración).
- **SECTOR**: clasifica a cada ocupado en sector formal (1) o informal (2), disponible en la EPH desde el 4T2024.

Se clasificó como **Informal self-employed** a quienes cumplen alguna de estas condiciones:
- Son cuenta propia (`CAT_OCUP==2`) y trabajan en el sector informal (`SECTOR==2`)
- Son trabajadores familiares sin remuneración (`CAT_OCUP==4`)

Los casos con valores faltantes en alguna de estas variables fueron excluidos del análisis.

```python
# --- 3.1 Construccion del indicador de informalidad ---
# Informal self-employed segun Maurizio & Monsalvo (2021):
# - Cuenta propia en sector informal (CAT_OCUP=2 y SECTOR=2)
# - Trabajador familiar sin remuneracion (CAT_OCUP=4)

def construir_informalidad(df):
    cond_informal = (
        ((df['CAT_OCUP'] == 2) & (df['SECTOR'] == 2)) |
        (df['CAT_OCUP'] == 4)
    )
    df = df.copy()
    df['informal_se'] = np.where(
        df['CAT_OCUP'].isna() | df['SECTOR'].isna(),
        np.nan,
        np.where(cond_informal, 1, 0)
    )
    return df

ocupados_2024 = construir_informalidad(ocupados_2024)
ocupados_2025 = construir_informalidad(ocupados_2025)

print('2024 - Distribucion informal_se:')
print(ocupados_2024['informal_se'].value_counts(dropna=False))
print('\n2025 - Distribucion informal_se:')
print(ocupados_2025['informal_se'].value_counts(dropna=False))
```

```python
# --- 3.2 Tabla doble entrada ---

def tabla_informalidad(df, anio):
    grupos = {'Total': df, 'Mujeres': df[df['Género'] == 1], 'Varones': df[df['Género'] == 0]}
    resultados = {}
    for nombre, subdf in grupos.items():
        subdf_valido = subdf[subdf['informal_se'].notna()]
        total_pond = subdf_valido['PONDERA'].sum()
        inf_pond   = subdf_valido.loc[subdf_valido['informal_se'] == 1, 'PONDERA'].sum()
        form_pond  = subdf_valido.loc[subdf_valido['informal_se'] == 0, 'PONDERA'].sum()
        resultados[nombre] = {
            'Informal self-employed (%)': round(inf_pond / total_pond * 100, 1),
            'Resto de ocupados (%)':      round(form_pond / total_pond * 100, 1),
            'N observaciones':            len(subdf_valido)
        }
    return pd.DataFrame(resultados).T.rename_axis(f'EPH {anio}T4')

tabla_2024 = tabla_informalidad(ocupados_2024, 2024)
tabla_2025 = tabla_informalidad(ocupados_2025, 2025)
tabla_final = pd.concat([tabla_2024, tabla_2025], keys=['2024T4', '2025T4'])

print('Tabla 3. Incidencia de Informal self-employed por sexo\n')
display(tabla_final)

# Exportar CSV
tabla_final.to_csv('tabla_informalidad.csv')
print('Exportado: tabla_informalidad.csv')

# Exportar PNG (principios Schwabish)
fig, ax = plt.subplots(figsize=(10, 4))
ax.axis('off')
tabla_reset = tabla_final.reset_index()
tabla_reset.columns = ['Año', 'Grupo', 'Informal SE (%)', 'Resto (%)', 'N obs.']
tbl = ax.table(cellText=tabla_reset.values, colLabels=tabla_reset.columns,
               loc='center', cellLoc='center')
tbl.auto_set_font_size(False)
tbl.set_fontsize(11)
tbl.scale(1.2, 1.8)
for (row, col), cell in tbl.get_celld().items():
    if row == 0:
        cell.set_facecolor('#2E75B6')
        cell.set_text_props(color='white', fontweight='bold')
    elif row % 2 == 0:
        cell.set_facecolor('#E8F0F8')
plt.title('Los varones presentan mayor incidencia de informalidad que las mujeres en ambos años\n'
          'Tabla 3. Incidencia de Informal self-employed por sexo y año',
          fontsize=12, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('tabla_informalidad.png', dpi=150, bbox_inches='tight')
plt.show()
print('Exportado: tabla_informalidad.png')
```

Los resultados muestran que la incidencia del trabajo informal por cuenta propia (informal self-employed) se mantuvo relativamente estable entre ambos períodos, pasando de 15,8% en el 4T2024 a 16,6% en el 4T2025 sobre el total de ocupados. En ambos años se observa una brecha por sexo: los varones presentan mayor incidencia de informalidad (16,8% en 2024 y 17,3% en 2025) respecto a las mujeres (14,6% y 15,7% respectivamente). Se refleja también un leve crecimiento interanual en la tasa de informalidad.

```python
# --- 3.3 Grafico 4 paneles (principios Schwabish 2014) ---
# 1. Titulo que cuenta la historia
# 2. Sin elementos distractores
# 3. Texto integrado

color_inf  = '#E07B39'
color_form = '#3A7EBF'

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle(
    'Los trabajadores informales por cuenta propia perciben ingresos sistematicamente menores\n'
    'EPH 2024T4 y 2025T4 (pesos a precios 2025T4)',
    fontsize=13, fontweight='bold', y=1.01
)

configs = [
    (axes[0, 0], 'P21',  'hist', 'Ingreso Ocupacion Principal (P21) - Histograma'),
    (axes[0, 1], 'P21',  'kde',  'Ingreso Ocupacion Principal (P21) - Kernel'),
    (axes[1, 0], 'P47T', 'hist', 'Ingreso Total Individual (P47T) - Histograma'),
    (axes[1, 1], 'P47T', 'kde',  'Ingreso Total Individual (P47T) - Kernel'),
]

for ax, var, tipo, titulo in configs:
    for df, anio, ls in [(ocupados_2024, '2024', '-'), (ocupados_2025, '2025', '--')]:
        inf  = df.loc[df['informal_se'] == 1, var].dropna()
        form = df.loc[df['informal_se'] == 0, var].dropna()
        if tipo == 'hist':
            ax.hist(inf,  bins=60, density=True, alpha=0.45, color=color_inf,  label=f'Informal SE {anio}')
            ax.hist(form, bins=60, density=True, alpha=0.45, color=color_form, label=f'Resto {anio}')
        else:
            inf.plot.kde(ax=ax,  color=color_inf,  linestyle=ls, linewidth=1.8, label=f'Informal SE {anio}')
            form.plot.kde(ax=ax, color=color_form, linestyle=ls, linewidth=1.8, label=f'Resto {anio}')
    ax.set_title(titulo, fontsize=11, fontweight='bold')
    ax.set_xlabel('Pesos (a precios 2025T4)', fontsize=9)
    ax.set_ylabel('Densidad', fontsize=9)
    ax.xaxis.set_major_formatter(
        ticker.FuncFormatter(lambda x, _: f'${x/1e6:.1f}M' if x >= 1e6 else f'${x/1e3:.0f}K')
    )
    ax.legend(fontsize=8, framealpha=0.7, title='Grupo')
    ax.spines[['top', 'right']].set_visible(False)
    ax.grid(axis='y', alpha=0.3, linewidth=0.5)

plt.tight_layout()
plt.savefig('distribucion_ingresos_informalidad.png', dpi=150, bbox_inches='tight')
plt.show()
print('Exportado: distribucion_ingresos_informalidad.png')
```

Los gráficos muestran que los trabajadores informales por cuenta propia perciben ingresos sistemáticamente menores que el resto de los ocupados. En ambos años y para ambas variables de ingreso, la distribución del grupo informal se concentra en los tramos bajos, con una cola derecha más corta, lo que indica que son pocos los trabajadores informales que alcanzan ingresos elevados.

Respecto a las limitaciones de las variables utilizadas, el **P21** captura únicamente el ingreso de la ocupación principal, por lo que subestima el ingreso real de quienes tienen más de un trabajo, situación habitual entre los trabajadores informales. El **P47T**, por su parte, incluye ingresos no laborales como jubilaciones, transferencias y programas sociales, lo que puede distorsionar la comparación entre grupos ya que no refleja exclusivamente lo que cada trabajador genera con su actividad laboral. A esto se suma que la tasa de no respuesta en ambas variables es mayor entre los informales, lo que introduce un sesgo en el análisis.
